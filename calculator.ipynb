```python
import tkinter as tk
from tkinter import ttk

# Function to update the input field
def update_input(value):
    current_input = entry_input.get()
    entry_input.delete(0, tk.END)
    entry_input.insert(0, current_input + value)

# Function to perform arithmetic operations
def calculate():
    try:
        expression = entry_input.get()
        result.set(eval(expression))
    except Exception as e:
        result.set("Invalid input")

# Function to clear the input field
def clear_input():
    entry_input.delete(0, tk.END)
    result.set("")

# Create the main window
window = tk.Tk()
window.title("Simple Calculator")

# Create and configure input widgets
entry_input = tk.Entry(window, width=20, font=("Arial", 18))
entry_input.grid(row=0, column=0, columnspan=4, padx=10, pady=10, ipadx=5, ipady=5)

# Create number and operator buttons
buttons = [
    '7', '8', '9', '+',
    '4', '5', '6', '-',
    '1', '2', '3', '*',
    '0', 'C', '=', '/',
]

row_num, col_num = 1, 0
for button_text in buttons:
    if button_text == '=':
        button = tk.Button(window, text=button_text, width=5, height=2, command=calculate)
    elif button_text == 'C':
        button = tk.Button(window, text=button_text, width=5, height=2, command=clear_input)
    else:
        button = tk.Button(window, text=button_text, width=5, height=2, command=lambda text=button_text: update_input(text))

    button.grid(row=row_num, column=col_num, padx=5, pady=5)
    col_num += 1
    if col_num > 3:
        col_num = 0
        row_num += 1

# Create a label to display the result
result = tk.StringVar()
result_label = tk.Label(window, textvariable=result, font=("Arial", 18))
result_label.grid(row=row_num, column=0, columnspan=4, padx=10, pady=10)

# Configure grid row and column weights to make the input field expandable
window.grid_rowconfigure(0, weight=1)
window.grid_columnconfigure(0, weight=1)

# Run the Tkinter event loop
window.mainloop()
```
