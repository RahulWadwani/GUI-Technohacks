```python
import tkinter as tk

class ATM:
    def __init__(self, root):
        self.root = root
        self.root.title("ATM Simulator")
        self.root.geometry("400x300")

        # Initialize user data (in-memory database)
        self.users = {
            "user1": {"password": "1234", "balance": 1000.00},
            "user2": {"password": "5678", "balance": 1500.00}
        }

        # Initialize user state
        self.current_user = None

        self.init_ui()

    def init_ui(self):
        # Create login frame
        self.login_frame = tk.Frame(self.root)
        self.login_frame.pack(pady=20)

        # Create username and password labels
        self.username_label = tk.Label(self.login_frame, text="Username:")
        self.password_label = tk.Label(self.login_frame, text="Password:")
        self.username_label.grid(row=0, column=0, padx=10, pady=10)
        self.password_label.grid(row=1, column=0, padx=10, pady=10)

        # Create username and password entry widgets
        self.username_entry = tk.Entry(self.login_frame)
        self.password_entry = tk.Entry(self.login_frame, show="*")
        self.username_entry.grid(row=0, column=1, padx=10, pady=10)
        self.password_entry.grid(row=1, column=1, padx=10, pady=10)

        # Create login button
        self.login_button = tk.Button(self.login_frame, text="Login", command=self.login)
        self.login_button.grid(row=2, columnspan=2, padx=10, pady=10)

    def login(self):
        username = self.username_entry.get()
        password = self.password_entry.get()

        if username in self.users and self.users[username]["password"] == password:
            self.current_user = username
            self.show_transactions_ui()
        else:
            tk.messagebox.showerror("Login Error", "Invalid username or password")

    def show_transactions_ui(self):
        # Remove the login frame
        self.login_frame.destroy()

        # Create transactions frame
        self.transactions_frame = tk.Frame(self.root)
        self.transactions_frame.pack(pady=20)

        # Create balance label
        self.balance_label = tk.Label(self.transactions_frame, text=f"Account Balance: ${self.users[self.current_user]['balance']:.2f}")
        self.balance_label.grid(row=0, column=0, padx=10, pady=10)

        # Create Withdraw and Deposit buttons
        self.withdraw_button = tk.Button(self.transactions_frame, text="Withdraw", command=self.withdraw)
        self.deposit_button = tk.Button(self.transactions_frame, text="Deposit", command=self.deposit)
        self.withdraw_button.grid(row=1, column=0, padx=10, pady=10)
        self.deposit_button.grid(row=1, column=1, padx=10, pady=10)

    def withdraw(self):
        withdrawal_amount = float(tk.simpledialog.askstring("Withdraw", "Enter withdrawal amount:"))
        if withdrawal_amount <= 0:
            tk.messagebox.showerror("Error", "Invalid withdrawal amount")
        elif withdrawal_amount > self.users[self.current_user]['balance']:
            tk.messagebox.showerror("Error", "Insufficient funds")
        else:
            self.users[self.current_user]['balance'] -= withdrawal_amount
            self.balance_label.config(text=f"Account Balance: ${self.users[self.current_user]['balance']:.2f}")

    def deposit(self):
        deposit_amount = float(tk.simpledialog.askstring("Deposit", "Enter deposit amount:"))
        if deposit_amount <= 0:
            tk.messagebox.showerror("Error", "Invalid deposit amount")
        else:
            self.users[self.current_user]['balance'] += deposit_amount
            self.balance_label.config(text=f"Account Balance: ${self.users[self.current_user]['balance']:.2f}")

if __name__ == "__main__":
    root = tk.Tk()
    atm = ATM(root)
    root.mainloop()
```
