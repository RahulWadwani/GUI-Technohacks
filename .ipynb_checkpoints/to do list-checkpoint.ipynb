```python
import tkinter as tk
from tkinter import messagebox
import time
from threading import Thread

def add_task():
    task = entry_task.get()
    if task:
        listbox_tasks.insert(tk.END, task)
        entry_task.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Please enter a task.")

def delete_task():
    try:
        selected_task_index = listbox_tasks.curselection()[0]
        listbox_tasks.delete(selected_task_index)
    except IndexError:
        messagebox.showwarning("Warning", "Please select a task to delete.")

def start_timer():
    try:
        selected_task_index = listbox_tasks.curselection()[0]
        selected_task = listbox_tasks.get(selected_task_index)
        countdown_time = int(entry_timer.get())

        def countdown():
            for i in range(countdown_time, 0, -1):
                timer_var.set(f"Time left: {i} seconds")
                time.sleep(1)
            timer_var.set("Time's up!")

        timer_thread = Thread(target=countdown)
        timer_thread.start()

    except IndexError:
        messagebox.showwarning("Warning", "Please select a task to start the timer.")

# Create the main window
window = tk.Tk()
window.title("To-Do List")

# Create and configure input widgets
entry_task = tk.Entry(window, width=30)
entry_task.grid(row=0, column=0, padx=10, pady=10)

entry_timer = tk.Entry(window, width=10)
entry_timer.grid(row=0, column=1, padx=10, pady=10)

# Create buttons for adding, deleting tasks, and starting the timer
add_button = tk.Button(window, text="Add Task", width=20, command=add_task)
add_button.grid(row=0, column=2, padx=10, pady=10)

delete_button = tk.Button(window, text="Delete Task", width=20, command=delete_task)
delete_button.grid(row=1, column=2, padx=10, pady=10)

start_timer_button = tk.Button(window, text="Start Timer", width=20, command=start_timer)
start_timer_button.grid(row=1, column=1, padx=10, pady=10)

# Create a listbox to display tasks
listbox_tasks = tk.Listbox(window, height=10, width=50)
listbox_tasks.grid(row=1, column=0, padx=10, pady=10)

# Create a label for the timer
timer_var = tk.StringVar()
timer_label = tk.Label(window, textvariable=timer_var)
timer_label.grid(row=2, column=0, columnspan=3, padx=10, pady=10)

# Run the Tkinter event loop
window.mainloop()
```
