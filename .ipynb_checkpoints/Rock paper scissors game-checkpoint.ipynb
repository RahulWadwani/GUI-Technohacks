```python
import tkinter as tk
from tkinter import messagebox
import random

class RockPaperScissorsGame:
    def __init__(self, root):
        self.root = root
        self.root.title("Rock, Paper, Scissors Game")
        self.root.geometry("400x300")

        self.score_user = 0
        self.score_computer = 0

        self.label_score = tk.Label(root, text=f"User: {self.score_user} | Computer: {self.score_computer}")
        self.label_score.pack(pady=10)

        self.button_frame = tk.Frame(root)
        self.button_frame.pack()

        self.buttons = ["Rock", "Paper", "Scissors", "Quit"]
        self.button_objects = []

        for text in self.buttons:
            button = tk.Button(self.button_frame, text=text, width=10, height=2, command=lambda t=text: self.play_game(t))
            self.button_objects.append(button)

        # Arrange buttons in a horizontal layout
        for i, button in enumerate(self.button_objects):
            button.grid(row=0, column=i, padx=10, pady=10)

        # Place the "Quit" button at the bottom center
        self.button_objects[-1].grid(row=1, column=2, padx=10, pady=20)

    def play_game(self, user_choice):
        if user_choice == "Quit":
            self.root.quit()
            return

        choices = ["Rock", "Paper", "Scissors"]
        computer_choice = random.choice(choices)

        winner = self.determine_winner(user_choice, computer_choice)

        if winner == "User":
            self.score_user += 5
        elif winner == "Computer":
            self.score_computer += 5

        self.label_score.config(text=f"User: {self.score_user} | Computer: {self.score_computer}")

        message = f"User chose: {user_choice}\nComputer chose: {computer_choice}\n{winner} wins!"
        if winner == "Tie":
            message = f"User chose: {user_choice}\nComputer chose: {computer_choice}\nIt's a tie!"

        messagebox.showinfo("Result", message)

    def determine_winner(self, user_choice, computer_choice):
        if user_choice == computer_choice:
            return "Tie"
        elif (
            (user_choice == "Rock" and computer_choice == "Scissors")
            or (user_choice == "Paper" and computer_choice == "Rock")
            or (user_choice == "Scissors" and computer_choice == "Paper")
        ):
            return "User"
        else:
            return "Computer"

if __name__ == "__main__":
    root = tk.Tk()
    game = RockPaperScissorsGame(root)
    root.mainloop()
```
